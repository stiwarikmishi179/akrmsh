```python
# ════════════════════════════════════════════════════════════════════════
#  v40 TRADING BOT — COLAB RUNNER
#
#  ✅ CORRECT WAY:
#     1. Open this file on your computer
#     2. Copy ALL contents (Ctrl+A, Ctrl+C)
#     3. Paste into a NEW Colab cell
#     4. Click the ▶ RUN button on that cell
#
#  ❌ WRONG WAY (causes all errors):
#     Do NOT run: !python colab_runner.py
#     Do NOT run: !python v40/colab_runner.py
#     Do NOT split into multiple cells
# ════════════════════════════════════════════════════════════════════════

# ── YOUR SETTINGS (only change these 4 things) ──────────────────────────
ZIP_FILENAME = "trading_bot_v40_testmode.zip"
TRADING_MODE = "paper"
TEST_MODE    = True    # True = trades in 5-10 min | False = production
MARKETS = {
    "crypto":      True,   # works without API key
    "stocks":      False,
    "forex":       False,
    "futures":     False,
    "commodities": False,
}
# ── END SETTINGS ────────────────────────────────────────────────────────

import os, sys, re, glob, time, shutil, zipfile, subprocess, threading

# ── Safety check: must be running as notebook cell, not subprocess ───────
def _is_notebook():
    try:
        from IPython import get_ipython
        return get_ipython() is not None
    except ImportError:
        return False

if not _is_notebook():
    print("=" * 60)
    print("ERROR: This file must be pasted into a Colab cell and run")
    print("from there. Do NOT run with !python colab_runner.py")
    print("See START_HERE.md for correct instructions.")
    print("=" * 60)
    import sys; sys.exit(1)

# ── Step 1: Mount Drive ──────────────────────────────────────────────────
from google.colab import drive, userdata
print("[1/9] Mounting Google Drive...")
try:
    drive.mount('/content/drive', force_remount=False)
    print("      ✅ Mounted")
except Exception as e:
    print(f"      ⚠️  Mount warning (may already be mounted): {e}")

# ── Step 2: Find zip ─────────────────────────────────────────────────────
print(f"[2/9] Looking for {ZIP_FILENAME}...")
found_zip = None
for base in ['/content/drive/MyDrive', '/content/drive/My Drive']:
    if not os.path.isdir(base):
        continue
    exact = os.path.join(base, ZIP_FILENAME)
    if os.path.exists(exact):
        found_zip = exact
        break
    hits = glob.glob(os.path.join(base, '**', ZIP_FILENAME), recursive=True)
    if hits:
        found_zip = hits[0]
        break
    # Fuzzy fallback
    for z in glob.glob(os.path.join(base, '**', '*.zip'), recursive=True):
        if 'v40' in z.lower():
            found_zip = z
            print(f"      ⚠️  Exact name not found — using: {os.path.basename(z)}")
            break
    if found_zip:
        break

if not found_zip:
    print(f"      ❌ Could not find '{ZIP_FILENAME}' in Google Drive")
    print("         Upload the zip to Drive and check ZIP_FILENAME matches exactly")
    raise FileNotFoundError(ZIP_FILENAME)
print(f"      ✅ {os.path.basename(found_zip)}")

# ── Step 3: Extract ──────────────────────────────────────────────────────
print("[3/9] Extracting zip...")
EXTRACT_DIR = '/content/trading_bot'
BOT_DIR     = os.path.join(EXTRACT_DIR, 'v40')
if os.path.exists(EXTRACT_DIR):
    shutil.rmtree(EXTRACT_DIR)
os.makedirs(EXTRACT_DIR, exist_ok=True)
with zipfile.ZipFile(found_zip, 'r') as z:
    z.extractall(EXTRACT_DIR)
    if not os.path.isdir(BOT_DIR):
        print(f"      ❌ No v40/ folder. Contents: {os.listdir(EXTRACT_DIR)}")
        raise FileNotFoundError(BOT_DIR)
print(f"      ✅ {BOT_DIR}")

# ── Step 4: Set working directory and Python path ────────────────────────
print("[4/9] Setting paths...")
os.chdir(BOT_DIR)
for p in [BOT_DIR, EXTRACT_DIR]:
    if p not in sys.path:
        sys.path.insert(0, p)
print(f"      ✅ CWD={BOT_DIR}")

# ── Step 5: Install dependencies ────────────────────────────────────────
print("[5/9] Installing dependencies (1-3 min)...")
req = os.path.join(BOT_DIR, 'requirements.txt')
result = subprocess.run(
    [sys.executable, '-m', 'pip', 'install', '-r', req,
     '-q', '--exists-action', 'i'],
    capture_output=True, text=True, cwd=BOT_DIR
)
errs = [l for l in result.stderr.splitlines()
        if 'ERROR' in l and 'warning' not in l.lower()]
if not errs:
    print("      ✅ All packages ready")
else:
    print(f"      ⚠️  {len(errs)} package error(s) — usually safe to continue:")
    for e in errs[:3]:
        print(f"         {e}")

# ── Step 6: Configure settings.yaml ─────────────────────────────────────
print("[6/9] Configuring settings...")
settings_path = os.path.join(BOT_DIR, 'config', 'settings.yaml')
with open(settings_path) as f:
    cfg = f.read()

    # Set mode
    for m in ['paper', 'live']:
        cfg = cfg.replace(f'mode: "{m}"', f'mode: "{TRADING_MODE}"')

    # Market toggles
    for market, enabled in MARKETS.items():
        flag = 'true' if enabled else 'false'
        cfg = re.sub(
            rf'(  {re.escape(market)}:\n(?:.*\n){{0,5}}?    enabled: )(?:true|false)',
            rf'\g<1>{flag}', cfg
        )

    with open(settings_path, 'w') as f:
        f.write(cfg)

    # Load API keys
    ngrok_token = None
    loaded = []
    for name, field in [
        ('BINANCE_API_KEY',    'api_key'),
        ('BINANCE_API_SECRET', 'api_secret'),
    ]:
        try:
            val = userdata.get(name)
            if val:
                cfg2 = open(settings_path).read()
                cfg2 = re.sub(rf'(    {field}: ")[^"]*(")', rf'\g<1>{val}\g<2>',
                               cfg2, count=1)
                open(settings_path, 'w').write(cfg2)
                loaded.append(name)
        except Exception:
            pass

    try:
        ngrok_token = userdata.get('NGROK_AUTH_TOKEN')
    except Exception:
        pass

    print(f"      ✅ mode={TRADING_MODE} | "
          f"markets={[m for m,v in MARKETS.items() if v]} | "
          f"test={TEST_MODE}")

# ── Step 7: Verify imports ───────────────────────────────────────────────
print("[7/9] Verifying imports...")
all_ok = True
for mod, cls in [
    ('config.loader',             'get_config'),
    ('strategies.strategy_engine','StrategyEngine'),
    ('risk.risk_manager',         'RiskManager'),
    ('bot',                       'TradingBotV40'),
]:
    try:
        obj = getattr(__import__(mod, fromlist=[cls]), cls)
        print(f"      ✅ {cls}")
    except Exception as e:
        print(f"      ❌ {cls}: {e}")
        all_ok = False

if not all_ok:
    print("\n      ❌ Import errors above — cannot start bot")
    print("         Most common fix: Runtime → Factory reset runtime → re-run")
    raise ImportError("Core import failed")

# ── Step 8: Start bot in background ─────────────────────────────────────
print("[8/9] Starting bot...")
bot_instance  = None
bot_exception = None

def _run_bot():
    global bot_instance, bot_exception
    try:
        os.chdir(BOT_DIR)
        if BOT_DIR not in sys.path:
            sys.path.insert(0, BOT_DIR)
        from bot import TradingBotV40
        bot_instance = TradingBotV40(mode=TRADING_MODE, test_mode=TEST_MODE)
        bot_instance.start(dashboard=True)
    except Exception as e:
        bot_exception = e
        import traceback; traceback.print_exc()

threading.Thread(target=_run_bot, daemon=True).start()

t0 = time.time()
for i in range(30):
    time.sleep(1)
    if bot_exception:
        print(f"\n      ❌ Bot failed to start: {bot_exception}")
        print("         See traceback above for details")
        raise RuntimeError(str(bot_exception))
    if bot_instance and getattr(bot_instance, '_running', False):
        print(f"      ✅ Bot running ({i+1}s)")
        break
    print(f"      Starting... {i+1}s", end='\r')
else:
    print("      ⏳ Still initialising — check logs below")

# ── Step 9: ngrok tunnel + status loop ──────────────────────────────────
print("[9/9] Setting up dashboard...")
dashboard_url = None
if ngrok_token:
    try:
        subprocess.run([sys.executable, '-m', 'pip', 'install', 'pyngrok', '-q'],
                       capture_output=True)
        from pyngrok import ngrok as _ngrok
        _ngrok.kill()
        _ngrok.set_auth_token(ngrok_token)
        time.sleep(2)
        port = getattr(bot_instance, '_dashboard_port', 8000) if bot_instance else 8000
        tunnel = _ngrok.connect(port)
        dashboard_url = tunnel.public_url
        print(f"      ✅ Dashboard: {dashboard_url}")
    except Exception as e:
        print(f"      ⚠️  ngrok: {e}")
else:
    print("      ⚠️  No NGROK_AUTH_TOKEN → add to Colab secrets for external dashboard")

print()
print("=" * 62)
if TEST_MODE:
    print("  ⚡ TEST MODE — first trades expected within 5-10 minutes")
    print("  Change TEST_MODE=False at the top for production")
else:
    print("  🏭 PRODUCTION MODE — trades on high-confidence signals")
if dashboard_url:
    print(f"  🌐 Dashboard: {dashboard_url}")
print("  Press: Runtime → Interrupt execution  to stop")
print("=" * 62)
print()

# Status updates every 60s
t0    = time.time()
tick  = 0
while True:
    time.sleep(60)
    tick += 1
    mins = int((time.time() - t0) / 60)

    if bot_exception:
        print(f"❌ Bot crashed: {bot_exception}")
        break

    if not bot_instance:
        print(f"[{mins}m] Still initialising...")
        continue

    try:
        s     = bot_instance.rm.stats()
        opens = sum(1 for p in bot_instance.rm.positions.values()
                    if not p.closed)
        parts = [
            f"[{mins}m]",
            f"Trades:{s.get('total_trades',0)}",
            f"Win:{s.get('win_rate',0)*100:.1f}%",
            f"PnL:{s.get('total_pnl',0):+.4f}",
            f"Open:{opens}",
        ]
        if getattr(bot_instance, '_evolving_engine', None):
            e = bot_instance._evolving_engine.stats()
            parts.append(f"Gen:{e['generation']} Genomes:{e['active']}")
        if getattr(bot_instance, '_compounder', None):
            c = bot_instance._compounder.summary()
            parts.append(f"Scale:{c.get('current_scale',1.0):.2f}x")
        if getattr(bot_instance, '_dynamic_risk', None):
            d = bot_instance._dynamic_risk.summary()
            parts.append(f"Risk:{d.get('state','normal')}")
        print(" | ".join(parts))
    except Exception as e:
        print(f"[{mins}m] status: {e}")
```
